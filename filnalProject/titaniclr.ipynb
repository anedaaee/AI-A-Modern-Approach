```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
trainData = pd.read_csv('/kaggle/input/titanic/train.csv')
AgeAvrg = int(trainData['Age'].mean())
trainData['Age'] = trainData['Age'].fillna(AgeAvrg)
trainData
```

```python
trainData.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
trainData
```

```python
trainData = pd.get_dummies(trainData, columns=['Sex', 'Embarked'])
trainData
```

```python
X = trainData.drop('Survived', axis=1) 
Y = trainData['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
            X, Y, test_size=0.1, random_state=42
)
X_train
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(X_train.values,Y_train.values)
```

```python
import matplotlib.pyplot as plt
Y_predict = reg.predict(X_test.values)
Y_predict_binary = [1 if predidct > 0.5 else 0 for predidct in Y_predict]
```

```python
from sklearn.metrics import r2_score

evaluate = r2_score(Y_test,Y_predict_binary)
evaluate
```
